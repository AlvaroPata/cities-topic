```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm_notebook
```

```python
link = "http://archive.urbantheorylab.net/publications/"
r = requests.get(link)
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
corpus = pd.DataFrame(columns=["title", "url", "date", "text"])
i = 0
for publication in soup.select('div.half.left-column-padding'):
    date = np.nan
    url = 'http://archive.urbantheorylab.net' + publication.select('h1>a')[0]['href']
    title = publication.find('h1').text
    paragraphs = [p.text for p in publication.find_all('p')]
    text = ' '.join(paragraphs).replace('DOWNLOAD', '').strip()
    corpus.loc[i, ['title', 'url', 'date', 'text']] = [title, url, date, text]
    i += 1
```
