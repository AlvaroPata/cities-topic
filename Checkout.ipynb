```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

# Prep

## Urls

```python
urls = pd.read_csv("links.csv", header=None)
```

```python
urls.columns = ['title', 'url']
```

```python
urls.title = urls.title.str.strip()
```

```python
urls
```

## Captchas

```python
for i, url in enumerate(urls.url):
    try:
        r = requests.get(url)
        html = r.text
        if "captcha" in html:
            urls.loc[i, 'captcha'] = True
            print(url, 'captcha')
        else:
            urls.loc[i, 'captcha'] = False
            print('no captcha')
    except Exception as e:
        urls.loc[i, 'captcha'] = e
        print(e)
        continue

urls
```

## Fine urls

```python
links = urls[urls.captcha == False]
links.reset_index(inplace=True, drop=True)
links.head()
```

```python
links[['title', 'url']].to_csv('sane_links.csv')
```

## Functions

```python
def print_output(url):
    r = requests.get(url)
    html = r.text
    soup = BeautifulSoup(html, 'html.parser')
    print(soup.prettify())
```
